```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun_dydx(x,r):
    dydx = -x / (r**2 - x**2)**0.5
    return dydx

def fun_kreis(x,r):
    kreis = (r**2 - x**2)**0.5
    return kreis
```

```python
x = np.arange(0,15e-3,1e-5)
dx = np.arange(0,14.9999999e-3,1e-5)
r = 15e-3

kreis = fun_kreis(x,r)
dydx = fun_dydx(dx,r)

global_angle = np.abs(np.arctan(dydx)*180/np.pi) # angle in degree

#global_angle = 1/np.arctan(dydx)*180/np.pi # angle in degree
#global_angle = np.arctan(dydx) # angle in radian
'''
Plot the data
'''
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))


ax[0].plot(x, kreis,"o:",color="tab:blue")
ax[1].plot(dx, dydx,"o:",color="tab:blue")
ax[2].plot(dx, global_angle,"o:",color="tab:blue")

ax[0].set_title('circle')
ax[1].set_title('deviation')
ax[2].set_title('contact angle')

ax[0].set_ylabel("y / -")
ax[1].set_ylabel("dydx / -")
ax[2].set_ylabel("angle / -")

ax[0].set_xlabel("x / -")
ax[1].set_xlabel("x / -")
ax[2].set_xlabel("x / -")

ax[0].set_ylim(0,20e-3)
ax[0].set_xlim(0,20e-3)

ax[2].set_ylim(0.1,95)
ax[2].set_xlim(0,16e-3)
plt.tight_layout()
plt.savefig("volume_refinment_validation.png")
```
