```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun_y(angle1 ,angle2 ,length ,y0 ,x):
    z1 = np.arcsinh(np.tan(-angle1))
    z2 = np.arcsinh(np.tan(angle2))

    z1 = np.arcsinh(-np.tan(angle1))
    z2 = np.arcsinh(np.tan(angle2))

    d = length

    a = -d / (z1 - z2)
    xsp = (z1 *d) / (z1 - z2)
    b = y0 - a * np.cosh(z1)

    yx = a *np.cosh((x-xsp)/a) + b
    return yx

def fun_y2(angle1 ,angle2 ,length ,yd ,x):
    z1 = np.arcsinh(np.tan(-angle1))
    z2 = np.arcsinh(np.tan(angle2))

    z1 = np.arcsinh(-np.tan(angle1))
    z2 = np.arcsinh(np.tan(angle2))

    d = length

    a = -d / (z1 - z2)
    xsp = (z1 *d) / (z1 - z2)
    b = y0 - a * np.cosh(z1)
    b = yd - a * np.cosh(z1-1)

    yx = a *np.cosh((x-xsp)/a) + b
    return yx
'''
def fun_diff_y():

    z1 = np.arcsinh(-np.tan(angle1))
    z2 = np.arcsinh(np.tan(angle2))

    a = -d / (z1 - z2)
    y_diff = a * np.cosh(z1-1) - a * np.osh(z1)
    return y_diff

'''
```

```python

angle1 = 0*np.pi /180
angle2 = 1e-6*np.pi /180
angle2 = 1*np.pi /180
#angle2 = 5*np.pi /180
#angle2 = 0*np.pi /180

length = 2
y0 = 1
x = np.arange(0,length,1e-6)

catenoid = fun_y(angle1, angle2, length,y0,x)
catenoid2 = fun_y(angle1, angle2, length,y0,x)


#global_angle = 90 + np.arctan(dydx)*180/np.pi # angle in degree
#global_angle = np.arctan(dydx) # angle in radian
'''
Plot the data
'''
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20, 6))


ax[0].plot(x, catenoid,"--",color="tab:blue")
ax[0].plot(x, catenoid2,"--",color="tab:orange")


ax[0].set_title('circle')
ax[1].set_title('deviation')
ax[2].set_title('contact angle')

ax[0].set_ylabel("y / -")
ax[1].set_ylabel("dydx / -")
ax[2].set_ylabel("angle / -")

ax[0].set_xlabel("x / -")
ax[1].set_xlabel("x / -")
ax[2].set_xlabel("x / -")


plt.tight_layout()
plt.savefig("catenoid_function.png")
```
