```python
# 1. Generate ellipsoid meshes and robustly visualize surface (solid blue faces, black edges, no points)

import numpy as np
import meshio
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
from _ellip_volume_pt1_Mesh_Quadrics import generate_meshes

def extract_surface_triangles_from_tets(tet_cells):
    """
    Returns list of surface triangles from tetrahedral cells:
    Triangles that appear only once (i.e. are not shared).
    """
    from collections import Counter
    faces = []
    for tet in tet_cells:
        for tri in [
            tuple(sorted([tet[0], tet[1], tet[2]])),
            tuple(sorted([tet[0], tet[1], tet[3]])),
            tuple(sorted([tet[0], tet[2], tet[3]])),
            tuple(sorted([tet[1], tet[2], tet[3]])),
        ]:
            faces.append(tri)
    face_counts = Counter(faces)
    surface_tris = [list(face) for face, count in face_counts.items() if count == 1]
    return surface_tris

ax, ay, az = 1.5, 1.0, 0.8

mesh_settings = [
    (0, 1.5, "Ellip_0"),   # Expect ~20 tets
    (1, 1.0, "Ellip_1"),   # Expect ~100 tets
    (2, 0.3, "Ellip_2"),   # Expect ~500 tets
    (3, 0.3, "Ellip_3"),   # Expect ~1000 tets
]

mesh_results = []
for subdivs, char_len, prefix in mesh_settings:
    print(f"Generating mesh: {prefix}")
    mesh_info = generate_meshes(ax, ay, az, subdivs, prefix, char_len)
    mesh_results.append(mesh_info)
    mesh_file = mesh_info['tet_meshname']
    mesh = meshio.read(mesh_file)
    points = mesh.points

    faces = []
    if 'triangle' in mesh.cells_dict and len(mesh.cells_dict['triangle']) > 0:
        tris = mesh.cells_dict['triangle']
        faces = [points[tri] for tri in tris]
        print(f"Found {len(faces)} surface triangles (from mesh).")
    elif 'tetra' in mesh.cells_dict and len(mesh.cells_dict['tetra']) > 0:
        tets = mesh.cells_dict['tetra']
        surface_tris = extract_surface_triangles_from_tets(tets)
        faces = [points[tri] for tri in surface_tris]
        print(f"Extracted {len(faces)} surface triangles (from tets).")
    else:
        print("No triangles or tetrahedra found! Can't plot this mesh.")
        continue

    if not faces:
        print("No surface faces found to plot for this mesh!")
        continue

    fig = plt.figure(figsize=(7, 5))
    ax3d = fig.add_subplot(111, projection='3d')
    ax3d.set_title(f"Ellipsoid mesh: {prefix}")

    poly = Poly3DCollection(
        faces,
        facecolor='deepskyblue',
        edgecolor='black',    # Strong black edges
        linewidths=1.0,       # Thicker lines for clarity
        alpha=0.85
    )
    ax3d.add_collection3d(poly)

    ax3d.set_xlabel('X')
    ax3d.set_ylabel('Y')
    ax3d.set_zlabel('Z')
    ax3d.auto_scale_xyz([points[:, 0].min(), points[:, 0].max()],
                        [points[:, 1].min(), points[:, 1].max()],
                        [points[:, 2].min(), points[:, 2].max()])
    plt.tight_layout()
    plt.show()

print("All ellipsoid meshes generated and visualized.")
```

```python
import pandas as pd
from _ellip_volume_pt2_Unify import unify_coeffs
from _volume_pt3_ComputeVolume import compute_volume_error

# Ellipsoid parameters (should match mesh generation)
ax, ay, az = 1.5, 1.0, 0.8
V_theory = 4/3 * 3.141592653589793 * ax * ay * az

# Adaptive patching parameters (should match mesh gen)
tol = 1e-4
max_depth = 6

SUMMARY_CSV = 'volume_error_summary.csv'
summary_rows = []

for mesh_info in mesh_results:
    print(f"Processing mesh: {mesh_info['tet_meshname']}")
    fit_csv_unified = unify_coeffs(mesh_info['quadric_csv'])
    summary = compute_volume_error(
        mesh_info['tet_meshname'],
        mesh_info['surf_meshname'],  # Not surf_from_tet_meshname!
        fit_csv_unified,
        V_theory,
        mesh_info.get('prefix', ''),  # Some scripts use a prefix, safe to include
        tol=tol,
        max_depth=max_depth
    )
    summary_rows.append(summary)
    # Save after each for progress recovery
    pd.DataFrame(summary_rows).to_csv(SUMMARY_CSV, index=False)
    print(f"Partial summary saved to {SUMMARY_CSV}")

print(f"All cases done. Final summary saved to {SUMMARY_CSV}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

csv_file = 'volume_error_summary.csv'
if not os.path.exists(csv_file) or os.stat(csv_file).st_size == 0:
    print(f"{csv_file} does not exist or is empty. Cannot plot.")
else:
    try:
        df = pd.read_csv(csv_file)
    except pd.errors.EmptyDataError:
        print(f"{csv_file} exists but has no columns. Cannot plot.")
    else:
        if df.empty or 'num_tet' not in df or 'rel_error_patchsum' not in df or 'elapsed_time_sec' not in df:
            print("CSV exists but does not have the expected columns. Cannot plot.")
        else:
            df = df.sort_values('num_tet')
            x = df['num_tet']
            y = df['rel_error_patchsum']
            t = df['elapsed_time_sec']

            fig, ax1 = plt.subplots(figsize=(7, 4.5))

            color = 'tab:blue'
            ax1.set_xlabel('Number of Tetrahedra (log scale)')
            ax1.set_ylabel('Relative Error (%) (log scale)', color=color)
            ax1.loglog(x, y, marker='o', color=color, linewidth=2)
            ax1.tick_params(axis='y', labelcolor=color)
            ax1.grid(True, which="both", ls='--', alpha=0.5)

            # Top x-axis for computing time, with 's' for seconds
            ax2 = ax1.twiny()
            ax2.set_xscale('log')
            ax2.set_xlim(ax1.get_xlim())
            ax2.set_xticks(x)
            ax2.set_xticklabels(['{:.1f} s'.format(val) for val in t], rotation=35, ha='left')
            ax2.set_xlabel('Computing Time (s, log scale)')

            plt.title('Relative Error vs Number of Tetrahedra (with Compute Time)')
            plt.tight_layout()
            plt.show()
```
