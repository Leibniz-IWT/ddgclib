```python
if __name__=="__main__":

    coeffs = (0.25, 1, -1, 0, 0, 0, 0, 0, 0, 0) # elliptic cone
    pt = (0.0, 0, 0.0) # elliptic cone


    pt_new = np.array(pt, float)

    Q = evaluate_quadric(coeffs, pt)
    print("### check if the point on the geometry surface before transforming ##########")   
   
    print("## ERROR #########",Q)    

    plot_quadric_surface(coeffs, "Original Quadric Surface")

    info = analyse_quadric(*coeffs,prescale=True, postscale=True)
    

    pt_new = info.R.T @ pt_new
    #print("pt_new after R", pt_new, "R", info.R)   

    pt_new = np.array(pt_new, float) / np.array(info.scale_factors1, float)
    #print("after scaling pt_new", pt_new, "scale_factors1", info.scale_factors1)
    #print("after scaling pt_new", pt_new, "scale_factors1",tuple(float(x) for x in info.scale_factors1))

    pt_new = pt_new - info.centre
    #print("pt_new after centre", pt_new, "centre", info.centre)
    Q2 = evaluate_quadric(info.new_coeffs, pt_new)
    print("### check if the point on the geometry surface after transforming ##########")   
    print("### ERROR ##########",Q2)    
    #detect_surface_of_revolution(info.new_coeffs)

    axis, name = detect_surface_of_revolution(info.new_coeffs)
    print("Old_coeffs", tuple(float(x) for x in coeffs))
    
    print("Axis:", axis)
    print("Geometry:", name)
    print("New_coeffs", tuple(float(x) for x in info.new_coeffs))
    print("Centre Transfer", tuple(float(-x) for x in info.centre))
    print("New Centre", [0,0,0])
    print("Scale Factor",tuple(float(x) for x in info.scale_factors1))

    
    plot_quadric_surface(info.new_coeffs, "Canonical Quadric Surface")    
```
