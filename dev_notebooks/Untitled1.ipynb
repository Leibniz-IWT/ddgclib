```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# define the parameters
u = np.linspace(0, 2.0 * np.pi, 100)  # longitudinal parameter
v = np.linspace(-10, 10, 100)  # latitudinal parameter
u, v = np.meshgrid(u, v)
c = 5  # parameter related to the catenary curve and its distance from the axis

# compute the unduloid coordinates
x = c * np.cosh(v/c) * np.cos(u)
y = c * np.cosh(v/c) * np.sin(u)
z = v - c * np.sinh(v/c)

# plot the unduloid
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b')

plt.show()
```
