```python
# Energy minimization outline:
# tauspan = np.linspace(0, 10, 200)  # timestep
tauspan = np.linspace(0, 1, 20)  # timestep
#tauspan = np.linspace(0, 2, 60)  # timestep
for tau in tauspan:
    # Loop through all interior vertices
    cHCV = copy.copy(HC.V)
    for v in cHCV:
        # exclude boundaries
        if v in set(boundary_bottom):
            continue
        # Define top boundary condition
        if v in set(boundary_top):
            continue
        # Compute curvature for each interior vertex
        HNdA_i = curvature(v)
        # Step in time
        df = -gamma * HNdA_i
        v_new = v.x_a - tau * df
        HC.V.move(v, tuple(v_new))
```
